```python
all_sales_nosum_2 = all_sales_nosum.rename(columns={"Value":"Number of residential property sales"})
long_median_2 =long_median.rename(columns={"Value":"Median property sale price"})
```

```python
all_sales_nosum_2 = all_sales_nosum_2.drop(["combination", "Value Type"],axis=1)
long_median_2 =long_median_2.drop(["combination", "Value Type"],axis=1) 
```

```python
wide_combined = all_sales_nosum_2.merge(long_median_2, how="left")
```

```python
wide_combined
```

```python
wide_combined.to_excel("wide_combined.xlsx")
```

```python
reduced_wide1 = wide_combined[(wide_combined["Year"]<2008)]
reduced_wide2 = wide_combined[(wide_combined["Year"]>2007)]
mins = reduced_wide2.groupby(["UK Region", "Property Type"],as_index=False).agg({
            "Number of residential property sales": np.min, 
            "Median property sale price" : np.min
        })
maxs = reduced_wide1.groupby(["UK Region", "Property Type"],as_index=False).agg({
            "Number of residential property sales": np.max, 
            "Median property sale price" : np.max
        })
```

```python
maxs = maxs.rename(columns={"Number of residential property sales":"pre_crash_max_sales", 
                    "Median property sale price":"pre_crash_max_price"})
mins = mins.rename(columns={"Number of residential property sales":"post_crash_min_sales", 
                    "Median property sale price":"post_crash_min_price"})
```

```python
joined_min_max = maxs.merge(mins, how="left")
```

```python
joined_min_max["sales_drop_frac"] = (joined_min_max["post_crash_min_sales"] / joined_min_max["pre_crash_max_sales"]) * 100
joined_min_max["price_drop_frac"] = (joined_min_max["post_crash_min_price"] / joined_min_max["pre_crash_max_price"]) * 100
```

```python
joined_min_max
```

```python
sales_drops_mat = joined_min_max[["UK Region", "Property Type", "sales_drop_frac"]]
```

```python
sales_drop_mat = sales_drops_mat.pivot(index="UK Region",columns="Property Type", values="sales_drop_frac")
```

```python
sns.heatmap(sales_drop_mat)
plt.title("Percentage of number of sales after financial crash")
```

```python
price_drops_mat = joined_min_max[["UK Region", "Property Type", "price_drop_frac"]]
price_drop_mat = price_drops_mat.pivot(index="UK Region",columns="Property Type", values="price_drop_frac")
```

```python
price_drop_mat
```

```python
sns.heatmap(price_drop_mat)
plt.title("Percentage of median sale price after financial crash")
```
